```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# import warnings
# warnings.filterwarnings('ignore')  #this will ignore the warnings.it wont display warnings in notebook

# Styling 

# plt.style.use("seaborn-bright")
# plt.rcParams['figure.facecolor'] = 'EFEFEF'
# plt.rcParams['axes.facecolor'] = 'EFEFEF'

plt.style.use('./daan.mplstyle')

# Get Data 
data = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names=["sepal_width", "petal_length", "petal_width", "sepal_length", "species"])


# Figure 1: filter by species
%matplotlib inline

sns.FacetGrid(data, hue='species', height=5).map(plt.scatter,'sepal_width','petal_length')
sns.FacetGrid(data, hue='species', height=5).map(plt.scatter,'sepal_width','petal_width')
sns.FacetGrid(data, hue='species', height=5).map(plt.scatter,'sepal_width','sepal_length')

sns.FacetGrid(data, hue='species', height=5).map(plt.scatter,'petal_length','petal_width')
sns.FacetGrid(data, hue='species', height=5).map(plt.scatter,'petal_length','sepal_length')

sns.FacetGrid(data, hue='species', height=5).map(plt.scatter,'petal_width','sepal_length')
```
