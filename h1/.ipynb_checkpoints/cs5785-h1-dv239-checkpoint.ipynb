```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set custom styling
plt.style.use('../daan.mplstyle')

# Get Data 
data = pd.read_csv('house-prices-advanced-regression-techniques/test.csv')


print(data.head())
```

# Part 1. The Housing Prices

### Give three examples of continous ant categorical features in the dataset

**Continous**

1. `YearBuilt`: the year the house was built
2. `YrSold`: the year the hous was sold
2. `LotArea`: the total area of the lot

**Categorical**

1. `Utilities`: indicates type of utilities 
2. `SaleCondition`: indicates condiation of house
3. `MSZoning`: indicates zoning type


```python
# Generate histogram of `YearsBuilt`

data["YearBuilt"].plot.hist(grid=True, bins=90, rwidth=0.9, color='#242424')

plt.title('Year house was built')
plt.xlabel('Year')
plt.ylabel('Amount of houses')
plt.grid(axis='y', alpha=0.75)
```

```python
# Generate histogram of `MSZoning`

groupedByZoning = data.groupby(["SaleCondition"]).size()
groupedByZoning = data["SaleCondition"].value_counts()

groupedByZoning.plot.bar(grid=True, bins=90, width=0.9, color='#242424')
#groupedByZoning.plot.hist(grid=True, bins=90 rwidth=0.9, color='#242424')

#plt.title('Year house was built')
#plt.xlabel('Year')
#plt.ylabel('Amount of houses')
#plt.grid(axis='y', alpha=0.75)

print(groupedByZoning.head())
```
